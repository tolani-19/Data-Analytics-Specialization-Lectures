```python
# Use df.columns command to drop columns via indexing
df2 = df.drop(df.columns[[0,3,8,9,12,13,14,15,16,17,22,23,24,25,28,29,30]], axis=1)
```

```python
df2.columns
```

```python
# filling missing values in the converted column
df2['Converted'] = df2['Converted'].replace(np.NaN, 'No') 
```

```python
# filling missing values in the converted column
df2['Converted'] = df2['Converted'].replace(np.NaN, 'No') 
```

```python
df2['Personal Interests'] = df2['Personal Interests'].str.replace(r'</p>|<p>|\n', '',regex = True)
```

```python
df2['Education'] = df2['Education'].replace('Bachelor&#039;s Degree', 'Bachelors')
```

```python
# creating an 'Age' column
def age(born):
    born = datetime.strptime(born, "%Y-%m-%d").date()
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))
  
df2['Age'] = df2['Date of Birth'].apply(age)
  
# display(df2)
```

## Overall Conversion rate

```python
df2.groupby('Converted').count()['Prospect_ID']
```

```python
# In order to visualize the above in a plot, you can use the following code:
ax = df2.groupby('Converted').count()['Prospect_ID'].plot(
kind='bar',
color='skyblue',
grid=True,
figsize=(10, 7),
title='Conversion Count'
)
ax.set_xlabel('Converted')
ax.set_ylabel('Count')
plt.show()
```

```python
# We can use the following code to get the percentage of customers who bought our course offerings
Percentage_Convtd = df2.groupby('Converted').count()['Prospect_ID']/df2.shape[0]*100

Percentage_Convtd
```

```python
#representing the above pivot table using a pie chart

Percentage_Convtd.plot(
kind='pie',
figsize=(15, 7),
startangle=90,
#subplots=True,
autopct=lambda x: '%0.1f%%' % x
)
plt.show()
```

## Conversion by programme of interest

```python
programme_conversion = pd.pivot_table(df2, values='Converted', index='Programme of Interest', columns='Conversion', aggfunc=len)
```

```python
programme_conversion
```

```python
# Visualizing the data above using a pie chart.
programme_conversion.plot(
kind='pie',
figsize=(15, 7),
startangle=90,
subplots=True,
legend=False,
fontsize=12,
autopct=lambda x: '%0.1f%%' % x
)
for ax in plt.gcf().axes:
    ax.legend(bbox_to_anchor=(0,1),loc=6)
plt.show()
```
